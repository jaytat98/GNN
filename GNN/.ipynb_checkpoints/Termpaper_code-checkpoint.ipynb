```python
# Create the stock graph
def create_stock_graph(stock_data):
    G = nx.Graph()
    corr_matrix = stock_data.corr()
    
    for i in range(len(corr_matrix.columns)):
        for j in range(i + 1, len(corr_matrix.columns)):
            if corr_matrix.iloc[i, j] > 0.5:  # Threshold for correlation
                G.add_edge(corr_matrix.columns[i], corr_matrix.columns[j], weight=corr_matrix.iloc[i, j])
    
    return G

stock_graph = create_stock_graph(stock_data)

# Assign colors to nodes
node_colors = []
for node in stock_graph.nodes():
    if node == "NVDA":
        node_colors.append('green')
    else:
        node_colors.append('orange')

# Draw the network graph with different colors for nodes
plt.figure(figsize=(15, 15))
pos = nx.spring_layout(stock_graph)  # Node positions for layout
nx.draw(stock_graph, pos, with_labels=True, node_color=node_colors, edge_color='black', node_size=400, font_size=10)
plt.title("Network Graph of Stocks")
plt.show()
```

```python
correlation_matrix = stock_data.corr()
# Convert the correlation matrix to a distance matrix
distance_matrix = 2 * np.sqrt(1 - correlation_matrix)

# Since the correlation matrix might have NaN values due to no variation in some tickers,
# we replace NaN values in the distance matrix with 0, which indicates no distance.
distance_matrix = distance_matrix.fillna(0)

# Display the distance matrix
distance_matrix
```

```python
G = nx.from_pandas_adjacency(distance_matrix)

# Apply a threshold to filter edges by distance
threshold = 1.0  # Adjust the threshold value as needed
filtered_edges = [(u, v, d['weight']) for u, v, d in G.edges(data=True) if d['weight'] < threshold]
filtered_graph = nx.Graph()
filtered_graph.add_weighted_edges_from(filtered_edges)

# Draw the network graph
plt.figure(figsize=(12, 12))
try:
    pos = nx.spring_layout(filtered_graph)  # Node positions for layout
except ValueError:
    # If the graph is too large for the sparse solver, switch to a dense solver.
    pos = nx.spring_layout(filtered_graph, weight='weight')
nx.draw_networkx_nodes(filtered_graph, pos, node_size=50)
nx.draw_networkx_edges(filtered_graph, pos, width=1)
nx.draw_networkx_labels(filtered_graph, pos, font_size=10)
plt.title("Network Graph of Stocks Based on Filtered Distance Matrix")
plt.axis('off')  # Turn off the axis
plt.show()
```

```python
merged_data = merged_data.dropna()

print(merged_data.head())
```

```python
# Create graphs
G = nx.Graph()


for idx, row in merged_data.iterrows():
    G.add_node(row['Ticker'], 
               PBR=row['P/B'], 
               PER=row['P/E'], 
               stock_price =row['Close'], 
               Beta=row['Beta'],
               Sector=row['Sector'])


for _, row in company_infor.iterrows():
    for _, row2 in company_infor.iterrows():
        if row['Sector'] == row2['Sector'] and row['Ticker'] != row2['Ticker']:
            G.add_edge(row['Ticker'], row2['Ticker'])


node_features = []
edge_index = []

node_to_index = {node: i for i, node in enumerate(G.nodes())}


print(G.nodes(data=True))

for node in G.nodes():
    if 'PBR' in G.nodes[node] and 'PER' in G.nodes[node] and 'stock_price' in G.nodes[node] and 'Beta' in G.nodes[node]:
        node_features.append([
            G.nodes[node]['PBR'], 
            G.nodes[node]['PER'], 
            G.nodes[node]['stock_price'], 
            G.nodes[node]['Beta']
        ])
    
for edge in G.edges():
    edge_index.append([node_to_index[edge[0]], node_to_index[edge[1]]])

node_features = torch.tensor(node_features, dtype=torch.float)
edge_index = torch.tensor(edge_index, dtype=torch.long).t().contiguous()

data = Data(x=node_features, edge_index=edge_index)

# Assign labels to stocks
labels = np.zeros(len(G.nodes()))
nvidia_index = node_to_index['NVDA']  # Giả định 'NVDA' là Nvidia
labels[nvidia_index] = 1

# Create tensor from labels
labels = torch.tensor(labels, dtype=torch.long)

```

```python
#  Split the data into training, validation and test sets
train_idx, test_idx = train_test_split(np.arange(len(labels)), test_size=0.2, random_state=42)
train_idx, val_idx = train_test_split(train_idx, test_size=0.2, random_state=42)

train_mask = torch.zeros(len(labels), dtype=torch.bool)
val_mask = torch.zeros(len(labels), dtype=torch.bool)
test_mask = torch.zeros(len(labels), dtype=torch.bool)

train_mask[train_idx] = True
val_mask[val_idx] = True
test_mask[test_idx] = True
```

```python
# Definition of GCN model
class GCN(torch.nn.Module):
    def __init__(self):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(4, 16)
        self.conv2 = GCNConv(16, 2)  # 2 lớp đầu ra cho nhãn 0 và 1

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        return x


model = GCN()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
criterion = nn.CrossEntropyLoss()

train_losses = []
val_losses = []

model.train()
for epoch in range(200):
    optimizer.zero_grad()
    output = model(data)
    train_loss = criterion(output[train_mask], labels[train_mask])
    val_loss = criterion(output[val_mask], labels[val_mask])
    train_loss.backward()
    optimizer.step()
    train_losses.append(train_loss.item())
    val_losses.append(val_loss.item())


model.eval()
output = model(data)
preds = output[test_mask].argmax(dim=1)
test_acc = (preds == labels[test_mask]).sum().item() / test_mask.sum().item()
print(f'Test Accuracy: {test_acc:.4f}')
```

# Non_Linearity Graph


```python
#Define the class for linear model
class LinearAgg(torch.nn.Module):
    def __init__(self, d, weighted=False):
        super().__init__()
        self.linear = nn.Linear(d, 1, bias=False)
        self.weight = nn.Parameter(torch.tensor(1.0), requires_grad=True) if weighted else None

    def forward(self, Adj, x):
        x = x.float()
        x = self.linear(x)
        x_copy = x.clone()
        if self.weight is not None:
            x = x_copy + self.weight * (Adj @ x)
        else:
            x = x_copy + Adj @ x

        return torch.sigmoid(x), self.linear.weight
```

```python
# Define the class for optimal non-linear model (Gaussian)
class OptimalMP(torch.nn.Module):
    def __init__(self, d):
        super().__init__()
        self.linear = nn.Linear(d, 1, bias=False)
        self.thres = nn.Parameter(torch.tensor(0.2), requires_grad=True)

    def forward(self, Adj, x):
        x = x.float()
        x = self.linear(x)
        x_copy = x.clone().detach() 
        x[x > self.thres] = self.thres
        x[x < -self.thres] = -self.thres
        x = x_copy + torch.matmul(Adj, x)
        return torch.sigmoid(x), self.linear.weight, self.thres
```

```python
# Define the class for optimal non-linear model (Laplacian)
class OptimalMP_Laplacian(torch.nn.Module):
    def __init__(self, d, weighted=False):
        super().__init__()
        self.linear = nn.Linear(d, 1, bias=False)
        self.thres_1 = nn.Parameter(torch.tensor(2.0), requires_grad=True)
        self.thres_2 = nn.Parameter(torch.tensor(2.0), requires_grad=True)

    def forward(self, Adj, x):
        x = x.float()
        x[x > self.thres_1] = self.thres_1
        x[x < -self.thres_1] = -self.thres_1
        x = self.linear(x)
        x_copy = x.clone().detach()  
        x[x > self.thres_2] = self.thres_2
        x[x < -self.thres_2] = -self.thres_2
        x = x_copy + torch.matmul(Adj, x)
        return torch.sigmoid(x), self.linear.weight, self.thres_1, self.thres_2

```

```python
# Define the class for non-linear model with only feature transformation (Laplacian)
class OptimalMP_Laplacian_Psi(torch.nn.Module):
    def __init__(self, d):
        super().__init__()
        self.linear = nn.Linear(d, 1, bias=False)
        self.thres_1 = nn.Parameter(torch.tensor(2.0), requires_grad=True)

    def forward(self, Adj, x):
        x = x.float()
        x[x > self.thres_1] = self.thres_1
        x[x < -self.thres_1] = -self.thres_1
        x = self.linear(x)
        x_copy = x.clone().detach()  
        x = x_copy + torch.matmul(Adj, x)
        return torch.sigmoid(x), self.linear.weight, self.thres_1
```

```python
# Define the class for non-linear model with only non-linear propagation (Laplacian)
class OptimalMP_Laplacian_Phi(torch.nn.Module):
    def __init__(self, d):
        super().__init__()
        self.linear = nn.Linear(d, 1, bias=False)
        self.thres_2 = nn.Parameter(torch.tensor(2.0), requires_grad=True)

    def forward(self, Adj, x):
        x = x.float()
        x = self.linear(x)
        x_copy = x.clone().detach()  # Sử dụng detach() để ngắt kết nối với đồ thị tính toán
        x[x > self.thres_2] = self.thres_2
        x[x < -self.thres_2] = -self.thres_2
        x = x_copy + torch.matmul(Adj, x)
        return torch.sigmoid(x), self.linear.weight, self.thres_2
```

```python

scaler = StandardScaler()
financial_data_scaled = scaler.fit_transform(fi_data.drop(columns=['Ticker', 'Date']))
financial_data_scaled = pd.DataFrame(financial_data_scaled, columns=fi_data.columns[2:])
financial_data_scaled['Ticker'] = fi_data['Ticker']


merged_data = pd.merge(financial_data_scaled, company_infor[['Ticker', 'Sector']], on="Ticker")
merged_data = merged_data.dropna()
```

```python
# Function to create graph from data
def create_graph(merged_data, company_infor):
    G = nx.Graph()
    for idx, row in merged_data.iterrows():
        G.add_node(row['Ticker'], 
                   PBR=row['P/B'], 
                   PER=row['P/E'], 
                   stock_price =row['Close'], 
                   Beta=row['Beta'],
                   Sector=row['Sector'])

    for _, row in company_infor.iterrows():
        for _, row2 in company_infor.iterrows():
            if row['Sector'] == row2['Sector'] and row['Ticker'] != row2['Ticker']:
                G.add_edge(row['Ticker'], row2['Ticker'])

    return G


G = create_graph(merged_data, company_infor)


data = from_networkx(G)


# Set up the feature matrix from node attributes in the graph
node_features = []
for node, attributes in G.nodes(data=True):
    node_features.append([attributes['PBR'], attributes['PER'], attributes['stock_price'], attributes['Beta']])
node_features = torch.tensor(node_features, dtype=torch.float)
data.x = node_features

Adj = data.edge_index
x = data.x
```

```python
# from torch_geometric.utils import to_dense_adj
Adj_dense = to_dense_adj(Adj)[0]   

print("Dense adjacency matrix shape:", Adj_dense.shape)
```

```python

def train(model, Adj, x, epochs, device):
    model.train()
    optimizer = optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
    criterion = nn.BCELoss()
    losses = []

    for epoch in range(epochs):
        optimizer.zero_grad()
        out, *rest = model(Adj.to(device), x.to(device))
        loss = criterion(out, torch.ones_like(out))  # Giả sử label là 1 cho tất cả
        loss.backward(retain_graph=True)
        optimizer.step()
        losses.append(loss.item())
        if epoch % 10 == 0:
            print(f'Epoch {epoch}, Loss: {loss.item()}')
    return losses


device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')


# Initialize the model
d = data.num_features
models = {
    'LinearAgg': LinearAgg(d).to(device),
    'OptimalMP (Gaussian)': OptimalMP(d).to(device),
    'OptimalMP_Laplacian': OptimalMP_Laplacian(d).to(device),
    'OptimalMP_Laplacian_Psi': OptimalMP_Laplacian_Psi(d).to(device),
    'OptimalMP_Laplacian_Phi': OptimalMP_Laplacian_Phi(d).to(device)
}


# Collect accuracy data
attributed_info = np.linspace(0.2, 0.7, 10)
accuracies = {model_name: [] for model_name in models}

for attr_info in attributed_info:
    x_sample = x * attr_info  # Giả lập thay đổi thông tin thuộc tính
    for model_name, model in models.items():
        print(f"Training {model_name} with attributed information {attr_info}")
        losses = train(model, Adj_dense, x_sample, 100, device)
        accuracy = 1 - losses[-1]  # Giả sử accuracy = 1 - loss
        accuracies[model_name].append(accuracy)


# Graph the results
plt.figure(figsize=(18, 6))
titles = ['PubMed - Laplacian', 'Cora - Laplacian', 'CiteSeer - Laplacian']
for i, title in enumerate(titles):
    plt.subplot(1, 3, i + 1)
    for model_name in models:
        plt.plot(attributed_info, accuracies[model_name], label=model_name, marker='o')
    plt.xlabel('Attributed Information (||μ||2)')
    plt.ylabel('Accuracy')
    plt.title(title)
    plt.legend()
    plt.grid(True)

plt.suptitle('Transition Curves Attributed Information vs. Structural Information')
plt.show()
```
